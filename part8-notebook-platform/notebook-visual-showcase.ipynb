## 3. 静态图表 — Matplotlib & Seaborn

### 2.4 中文字体配置

Matplotlib 默认字体不支持中文，需要先配置中文字体。以下代码自动检测系统中可用的中文字体（macOS/Windows/Linux），找不到时给出安装提示。

````python
# 2.4 Matplotlib Font Config (must run before importing pyplot)
import matplotlib
import matplotlib.font_manager as fm

# Try multiple Chinese fonts by priority
chinese_fonts = [
    'PingFang SC',        # macOS
    'Heiti SC',            # macOS
    'STHeiti',             # macOS
    'SimHei',              # Windows / common Linux
    'Microsoft YaHei',     # Windows
    'WenQuanYi Micro Hei', # Linux
    'Noto Sans CJK SC',    # Linux (Google)
    'Source Han Sans SC',  # Linux (Adobe)
    'Arial Unicode MS',    # macOS fallback
]

available_fonts = set(f.name for f in fm.fontManager.ttflist)
selected_font = None
for font_name in chinese_fonts:
    if font_name in available_fonts:
        selected_font = font_name
        break

if selected_font:
    matplotlib.rcParams['font.sans-serif'] = [selected_font, 'DejaVu Sans']
    print(f'✅ Chinese font set: {selected_font}')
else:
    # Font not found, try loading system font files dynamically
    import os, glob
    font_paths = []
    for pattern in [
        '/System/Library/Fonts/PingFang.ttc',           # macOS
        '/System/Library/Fonts/STHeiti*.ttf',            # macOS
        '/usr/share/fonts/**/wqy-microhei.ttc',          # Linux WenQuanYi
        '/usr/share/fonts/**/NotoSansCJK*.ttf',           # Linux Noto
        '/usr/share/fonts/**/SimHei*.ttf',                # Linux SimHei
        'C:/Windows/Fonts/simhei.ttf',                    # Windows
        'C:/Windows/Fonts/msyh.ttc',                      # Windows
    ]:
        font_paths.extend(glob.glob(pattern, recursive=True))
    if font_paths:
        fm.fontManager.addfont(font_paths[0])
        font_prop = fm.FontProperties(fname=font_paths[0])
        matplotlib.rcParams['font.sans-serif'] = [font_prop.get_name(), 'DejaVu Sans']
        print(f'✅ Chinese font loaded: {font_prop.get_name()} ({font_paths[0]})')
    else:
        print('⚠️ No Chinese font found! Install one:')
        print('  macOS: PingFang SC is built-in on macOS')
        print('  Linux: apt-get install fonts-wqy-microhei or pip install matplotlib')
        print('  Or download font and load manually: fm.fontManager.addfont("path/to/font.ttf")')
        print('  Workaround: use English titles')
        matplotlib.rcParams['font.sans-serif'] = ['DejaVu Sans']

# Fix minus sign display
matplotlib.rcParams['axes.unicode_minus'] = False
````

## 6. 交互式控件 — ipywidgets

````python
import ipywidgets as widgets
from ipywidgets import interact, interactive, FloatSlider, IntSlider, Dropdown, ToggleButtons

# 6.1 Interactive sine wave
@interact(
    freq=FloatSlider(min=0.1, max=10, step=0.1, value=2, description='Freq'),
    amp=FloatSlider(min=0.1, max=5, step=0.1, value=1, description='Amplitude'),
    phase=FloatSlider(min=0, max=6.28, step=0.1, value=0, description='Phase'),
    color=Dropdown(options=['#4facfe', '#f5576c', '#43e97b', '#fa709a'], value='#4facfe', description='Color'),
)
def plot_wave(freq, amp, phase, color):
    x = np.linspace(0, 10, 500)
    y = amp * np.sin(freq * x + phase)
    
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(x, y, linewidth=2.5, color=color)
    ax.fill_between(x, y, alpha=0.15, color=color)
    ax.set_xlim(0, 10)
    ax.set_ylim(-5.5, 5.5)
    ax.set_title(f'Sine wave: y = {amp:.1f} * sin({freq:.1f}x + {phase:.1f})', fontsize=14)
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='white', linewidth=0.5)
    plt.tight_layout()
    plt.show()
````

````python
import networkx as nx

# 9.1 Network topology
G = nx.karate_club_graph()

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Spring layout
pos1 = nx.spring_layout(G, seed=42)
degrees = dict(G.degree())
node_sizes = [v * 80 for v in degrees.values()]
colors = [G.nodes[n]['club'] == 'Mr. Hi' for n in G.nodes]
color_map = ['#4facfe' if c else '#f5576c' for c in colors]

nx.draw(G, pos1, ax=axes[0], with_labels=True, node_color=color_map,
        node_size=node_sizes, font_size=8, font_color='white',
        edge_color='gray', alpha=0.8, width=1.5)
axes[0].set_title('Karate Club Network (Spring Layout)', fontsize=14, fontweight='bold')

# Circular layout
pos2 = nx.circular_layout(G)
nx.draw(G, pos2, ax=axes[1], with_labels=True, node_color=color_map,
        node_size=node_sizes, font_size=8, font_color='white',
        edge_color='gray', alpha=0.6, width=1)
axes[1].set_title('Karate Club Network (Circular Layout)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print(f'Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}')
````

## 12. 生成图片与音频

````python
# 12.1 Generate image with Matplotlib
fig, ax = plt.subplots(figsize=(8, 4))
gradient = np.linspace(0, 1, 256).reshape(1, -1)
gradient = np.vstack([gradient] * 50)
for i, cmap_name in enumerate(['viridis', 'plasma', 'inferno', 'magma', 'coolwarm', 'Spectral']):
    ax.imshow(gradient, aspect='auto', cmap=cmap_name, extent=[0, 1, i*0.15, (i+1)*0.15])
    ax.text(-0.02, i*0.15 + 0.075, cmap_name, ha='right', va='center', fontsize=10, color='white')
ax.set_xlim(-0.15, 1)
ax.set_ylim(0, 6*0.15)
ax.set_title('Colormap Showcase', fontsize=14, color='white')
ax.axis('off')
plt.tight_layout()
plt.show()
````

````python
# 12.2 Generate and play audio waveform
sample_rate = 22050
duration = 2.0
t = np.linspace(0, duration, int(sample_rate * duration))

# Generate C major chord (C4 + E4 + G4)
c4 = 0.3 * np.sin(2 * np.pi * 261.63 * t)
e4 = 0.2 * np.sin(2 * np.pi * 329.63 * t)
g4 = 0.2 * np.sin(2 * np.pi * 392.00 * t)
audio_signal = c4 + e4 + g4

# Fade in/out envelope
envelope = np.ones_like(t)
fade_samples = int(0.1 * sample_rate)
envelope[:fade_samples] = np.linspace(0, 1, fade_samples)
envelope[-fade_samples:] = np.linspace(1, 0, fade_samples)
audio_signal *= envelope

# Show waveform
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(t[:1000], audio_signal[:1000], linewidth=1, color='#4facfe')
ax.set_title('C Major Chord Waveform (first 1000 samples)', fontsize=12)
ax.set_xlabel('Time (s)'); ax.set_ylabel('Amplitude')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Play audio
display(Audio(audio_signal, rate=sample_rate))
````
